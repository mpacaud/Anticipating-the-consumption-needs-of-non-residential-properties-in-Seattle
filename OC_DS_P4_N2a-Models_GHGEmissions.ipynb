## 2) Categorizing the output feature for a better stratification when splitting the dataset in a test set and a train set.

```python
# Name of the feature to predict and categorize.
output_feature_name = 'TotalGHGEmissions'
output_feature_name_categorized = output_feature_name + '_Categorized'
```

```python
def interquartile (serie):
    
    """
    Description
    -----------
    Calculate the interquartile.
    
    Parameters
    ----------
    serie: pandas.Serie
        Serie in which calculate the interquartile.
        
    Return: int
    ------
    Return the interquarile.
        
    """   
    
    return int(serie.quantile(q=0.75) - serie.quantile(q=0.25))
```

```python
def quantile_outlier_limit (serie):
    
    """
    Description
    -----------
    Calculate the Q1 - 1.5 * IQ and Q3 + 1.5 * IQ quantiles which are commonly use to define a limit over which
    values can be considered as uncommon or outliers.
    
    Parameters
    ----------
    serie: pandas.Serie
        Serie in which calculate the interquartile.
        
    Return: list
    ------
    Return the two values calculated packed in a list.
    
    """    
    
    qol_high = int(serie.quantile(q=0.75) + 1.5 * interquartile(serie))
    qol_low = int(serie.quantile(q=0.25) - 1.5 * interquartile(serie))
            
    return [qol_high, qol_low]
```

```python
%%time

# Use un simplified name for convinience.
df = data_model

# Set the figure environment.
fig = plt.figure(figsize=(8, 5), dpi=150)

# Draw the explicated features values.
plt.hist(df[[output_feature_name]], bins=1000)

# Calculate the common quantiles.
q1 = int(df[[output_feature_name]].quantile(q=0.25, axis=0))
q2 = int(df[[output_feature_name]].quantile(q=0.5, axis=0))
q3 = int(df[[output_feature_name]].quantile(q=0.75, axis=0))
q_low_limit = quantile_outlier_limit(df[[output_feature_name]])[1]
q_high_limit = quantile_outlier_limit(df[[output_feature_name]])[0]

# Display the quantiles to the graphic.
plt.axvline(q2, 0, 1, color='black')
plt.axvline(q1, 0, 1, color='red')
plt.axvline(q3, 0, 1, color='blue')
plt.axvline(q_high_limit, 0, 1, color='purple')
plt.axvline(q_low_limit, 0, 1, color='orange')

# Set axis parameters
plt.xlabel(output_feature_name)
plt.ylabel('Number of properties')
plt.xlim(-200, 1000)

# Display the figure.
plt.show()
```

```python
%%time

# Add a column to categorize the output feature in order to allow to starify the train and test split and the cross validations.
bins = [-np.inf, 2, 10, 100, 1000, np.inf] #bins=[-np.inf, q1, q2, q3, 500, np.inf]
labels=[1,2,3,4,5]
data_model[output_feature_name_categorized] = pd.cut(x = data_model[output_feature_name], 
                                                     bins=bins,
                                                     labels=labels)

# Check the categorization results.
display(data_model[[output_feature_name, output_feature_name_categorized]])

# Check if there enough properties in each categories.
display(data_model.groupby(output_feature_name_categorized).count()[[output_feature_name]])

# Check the most uncommon properties categories.
idxs = data_model[(data_model[output_feature_name] <= 0) | (data_model[output_feature_name] >= 5000)].index
data_raw.iloc[idxs][['PropertyName', 'PrimaryPropertyType', output_feature_name]]
```

### a) Investigation of optimal number of folds for cross-validations

```python
def kfold_tune (model, X, y, kmin=2, kmax=100):
    
    """
    Description
    -----------
    Find the best number of folds (k) within the defined range.
    
    Parameters
    ----------
    model: sklearn model
        Model to test.
    X: pandas.Dataframe()
        Dataframe of the explicatives.
    y: pandas.Dataframe()
        Dataframe of the explicated.
    kmin: int
        Lowest k to try.
    kmax: int
        Highest k to try.
        
    Return: int 
    ------
    Returns the best k found.
        
    """    
    
    # Check many k values on the error of many predicted values.
    cvs_means = []
    k_range = range(kmin, kmax)
    for k in k_range:
        cvs_means.append(cross_val_score(model, X, y, cv=k, n_jobs = -1).mean())
        if cvs_means[-1] >= max(cvs_means):
            k_best = k
        
    # Show the mean and standard deviation of the R2 error corresponding to each KFold shuffle at the defined random_state.
    print("Mean R2 =", np.array(cvs_means).mean(), "for KFold split numbers k tried in a range of [", kmin, ",", kmax, "].")
    print("Standard deviation of the mean R2 =", np.array(cvs_means).std(), "for KFold split numbers k tried in a range of [", kmin, ",", kmax, "].")
    print("Best k =", k_best, "for KFold split numbers k tried in a range of [", kmin, ",", kmax, "].")
    
    # Display the correpsonding graphic.
    plt.scatter(k_range, cvs_means)
    
    return k_best
```

### b) Hyperparameters tuning

```python
def hyperpara_tune (model, X, y, k, para_grid, rand_state=0):
    
    """
    Description
    -----------
    Tune the chosen hyperapameters of the model by checking all hyperparmateres combinations with the GridSearchCV() method.
    
    Parameters
    ----------
    model: sklearn model
        Model to test.
    X: pandas.Dataframe()
        Dataframe of the explicatives.
    y: pandas.Dataframe()
        Dataframe of the explicated.
    k: int
        Split parameter of the KFold function.
    para_grid: dictionary
        Parameters of the model to tune.
    rand_state: int
        random_state parameter in order to fix the randomness of the runs.
        
    Return: sklearn.GridSearchCV 
    ------
    Returns the fitted grid.
        
    """     
    
    # Set the KFold cross validation with the selected k.
    cv = KFold(k, shuffle=True, random_state=rand_state)

    # Search the best parameters among all parameters proposed.
    model_grid = GridSearchCV(estimator=model, param_grid=para_grid, cv=cv, n_jobs = -1, refit=True)

    # Train the grid search model.
    model_grid.fit(X, y.to_numpy().ravel())

    # Display the best hyperparameter.
    print("\n The best estimator across all searched params:\n", model_grid.best_estimator_)
    print("\n The best score across all searched params:\n", model_grid.best_score_)
    print("\n The best parameters across all searched params:\n", model_grid.best_params_)  
    
    return model_grid
```

```python
def hyperpara_tune_v4 (model, X, y, k, para_grid, n_iter=100, grid_loop=1, n_divs=10, range_precision=2,
                       int_para_names=['n_estimators', 'max_depth', 'min_samples_split', 'min_samples_leaf'],
                       bool_para_names=[True, False],
                       rand_state=0):
    
    """
    Description
    -----------
    Tune the chosen hyperapameters of the model by checking a random number of combinations with the RandomizedSearchCV() method.
    This function tunes the hyperparameters by taking the best combination among the best one found in each loop run.
    
    Parameters
    ----------
    model: sklearn model
        Model to test.
    X: pandas.Dataframe()
        Dataframe of the explicatives.
    y: pandas.Dataframe()
        Dataframe of the explicated.
    k: int
        Split parameter of the KFold function.
    para_grid: dictionary
        Parameters of the model to tune.
    n_iter: int
        Number of combinations of hyperparameters within which the RandomizedSearchCV will pick the best.
    grid_loop:
        Number of loop run.
    range_precision: int
        Number of decimals to round to.
    int_para_names: list of strings
        Hyperparameters which accept only integers.
    bool_para_names: list of strings
        Hyperparameters which accept only booleans.
    rand_state: int
        random_state parameter in order to fix the randomness of the runs.

    Return: sklearn.RandomizedSearchCV()
    ------
    Returns the fitted randomized model grid.
        
    """      
    
    def show ():
        
        """
        Description
        -----------
        Show relevant information at the end of the process.
        
        """  
        
        # Fit the grid model.
        model_grid.fit(X, y)
        
        # Display the best hyperparameter.
        print("\n The best estimator across all searched params:\n", model_grid.best_estimator_)
        print("\n The best score across all searched params:\n", model_grid.best_score_)
        print("\n The best parameters across all searched params:\n", model_grid.best_params_)  
    
    
    # Set the KFold cross validation with the selected k.
    cv = KFold(k, shuffle=True, random_state=rand_state)

    # Create a dictionary with all parameters to test as keys and empty lists as values.
    best_para_stored = para_grid.copy()
    for key in best_para_stored.keys():
        best_para_stored[key] = []
    
    for i in range(grid_loop):
        
        # Random search of parameters, use all available cores.
        model_grid = RandomizedSearchCV(estimator=model, param_distributions=para_grid, cv=cv, n_iter=n_iter, refit=True,
                                        n_jobs = -1, random_state=rand_state)
         
        # Fit the random grid.
        # NB: Step needed to be able to get the "best_params_" method.
        model_grid.fit(X, y.to_numpy().ravel())

        # Get the best parameters values in a dictionary.
        best_para = model_grid.best_params_
        
        # Loop to store the best parameter got in this loop run, in order to make the average at the end of all runs.
        for key in best_para_stored.keys():
            
            # Get the best value for the "key" parameter.
            best_para_value = round(best_para.get(key), range_precision)
            
            # Store this value in the dictionary set at the beginning of the function ("best_para_stored").
            best_para_stored[key].append(best_para_value)
            
            # Remove duplicates.
            best_para_stored[key] = list(set(best_para_stored[key]))
    
    #print(best_para_stored)
    # Replace the initial parameters with the best ones found.
    para_grid = best_para_stored
    
    # Find the best parameter among the best found.
    model_grid = GridSearchCV(estimator=model, param_grid=para_grid, n_jobs = -1)
    
    # Show results.
    show()
        
    return model_grid
```

### c) Evaluation of model performances

```python
def evaluate (model, X, y, n_folds=5, dataset='dataset', rand_state=None):
    
    """
    Description
    -----------
    Evaluate the performance of the model.
    
    Parameters
    ----------
    model: sklearn model
        Model to test.
    X: pandas.Dataframe()
        Dataframe of the explicatives.
    y: pandas.Dataframe()
        Dataframe of the explicated.
    n_folds: int
        Number of folds for the cross validation.
    data_set: str
        Display title for the dataset (ex: trainset or testset).
    rand_state: int
        Set the random_state for the cross validations.
        
    Return: list
    ------
    Return the score and its sandard deviation in a list.
        
    """
    
    if dataset == 'trainset':
        
        # Generate the predicted dataset generated by the trained model chosen.
        y_pred = cross_val_predict(model, X, y[output_feature_name].ravel(), cv=n_folds)
    
        # Error: Comparison between the cross validated predicted values and the original ones.
        print("- Errors on the", dataset + ":")
        print("Cross validation mean absolute error (MAE) on the " + dataset + ": {:0.4f}".format(mean_absolute_error(y[output_feature_name], y_pred)))
        print("Cross validation root mean squared error (RMSE) on the " + dataset + ": {:0.4f}".format(np.sqrt(mean_squared_error(y[output_feature_name], y_pred))))
        print("Cross validation median absolute error on the " + dataset + ": {:0.4f}\n".format(median_absolute_error(y[output_feature_name], y_pred)))
    
        # Model's performance measured by a simple score (R2) on the whole dataset at once.
        #print("R2 score on the", dataset, "with the given split: {:0.2f} %".format(model.score(X, y[output_feature_name]) * 100))

        # Model's performance measured by a KFold cross validation score (R2).
        multi_score = cross_val_score(model, X, y[output_feature_name], scoring='r2', cv=n_folds, n_jobs = -1)
        print("R2 score on the cross validated", dataset, ": {:0.2f}".format(multi_score.mean() * 100),
              "+/- {:0.2f} %".format(multi_score.std() * 100))
    
        # Model's performance measured by a stratified cross validation score (R2).
        #strat_cvs = cross_val_score_stratified_continuous(model, X, y, n_folds, rand_state=rand_state)
        #print("R2 stratified cross validation score: {:0.2f}".format(strat_cvs[1] * 100),
        #      "+/- {:0.2f} %".format(strat_cvs[2] * 100))
        
        # Add an empty line to the results display.
        print("")
        
        # Score to return.
        multi_score = cross_val_score(model, X, y[output_feature_name], scoring='r2', cv=n_folds, n_jobs = -1)
        score = round(multi_score.mean() * 100, 2)
        score_std = round(multi_score.std() * 100, 2)
        
        return [score, score_std]
    
    else:
    
        # Generate the predicted dataset generated by the trained model chosen.
        y_pred = model.predict(X)#, y[output_feature_name].ravel())
        
        # Error: Comparison between the cross validated predicted values and the original ones.
        print("- Errors on the", dataset + ":")
        print("Mean absolute error (MAE) on the " + dataset + ": {:0.4f}".format(mean_absolute_error(y[output_feature_name], y_pred)))
        print("Root mean squared error (RMSE) on the " + dataset + ": {:0.4f}".format(np.sqrt(mean_squared_error(y[output_feature_name], y_pred))))
        print("Median absolute error on the " + dataset + ": {:0.4f}\n".format(median_absolute_error(y[output_feature_name], y_pred)))
    
        # Model's performance measured by a simple score (R2) on the whole dataset at once.
        score = round(model.score(X, y[output_feature_name]) * 100, 2)
        print("R2 score on the", dataset, "with the given split: {:0.2f} %".format(score))
        
        # Add an empty line to the results display.
        print("")
        
        return [score]
```

### d) Visualization of errors

```python
def abs_error_hist (values, values_pred, bins, xlim = None, title = 'TITLE', err_label = "ABSCISSA"):
    
    """
    Description
    -----------
    Draw graphics showing the errors for more details on them.
    
    Parameters
    ----------
    values: np.array or pandas.DataFrame
        Values to predict.
    values_pred: list, np.array or pandas.DataFrame
        Predicted values.
    bins: int
        Number of splits for the histogram.
    xmin: int or float
        Max value of the abscissa to show on the graphic.
    title: str
        Figure title.
    err_label: str
        Figure abscissa label.
        
    """
    
    # Calculate the absolute error.
    errors = np.abs(values - values_pred)
    plt.axvline(mean_absolute_error(values_pred, values), 0, 1, color='red')
    plt.axvline(median_absolute_error(values_pred, values), 0, 1, color='green')
    
    # Generate the histogram in percentage.
    plt.hist(errors, bins=bins, weights=np.ones(len(errors)) / len(errors)*100)
    
    # Set some graph parameters
    plt.title(title)
    plt.xlabel(err_label)
    plt.ylabel('Proportion of errors (%)')
    plt.xlim(xlim)
    
    # Draw the histogram.
    plt.show()
```

```python
def root_squared_error_hist (values, values_pred, bins, xlim = None, title = 'TITLE', err_label = "ABSCISSA"):
    
    """
    Description
    -----------
    Draw graphics showing the errors for more details on them.
    
    Parameters
    ----------
    values: np.array or pandas.DataFrame
        Values to predict.
    values_pred: list, np.array or pandas.DataFrame
        Predicted values.
    bins: int
        Number of splits for the histogram.
    xmin: int or float
        Max value of the abscissa to show on the graphic.
    title: str
        Figure title.
    err_label: str
        Figure abscissa label.
        
    """
    
    # Calculate the absolute error.
    errors = np.sqrt(np.square(values - values_pred))
    plt.axvline(np.sqrt(mean_squared_error(values_pred, values)), 0, 1, color='red')
    
    # Generate the histogram in percentage.
    plt.hist(errors, bins=bins, weights=np.ones(len(errors)) / len(errors)*100)
    
    # Set some graph parameters
    plt.title(title)
    plt.xlabel(err_label)
    plt.ylabel('Proportion of errors (%)') 
    plt.xlim(xlim)
    
    # Draw the histogram.
    plt.show()
```
